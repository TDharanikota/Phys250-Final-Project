```python
def import_img(path):
    img = cv2.imread(path)
    rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    return rgb_img

def make_bw(img):
    bw_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    return bw_img

def change_bright(img, value):
    hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
    h, s, v = cv2.split(hsv)

    lim = 255 - value
    v[v > lim] = 255
    v[v <= lim] = np.add(v[v <= lim], value, out=v[v <= lim], casting='unsafe') # have to do np.add b/c += will not accept neg

    final_hsv = cv2.merge((h, s, v))
    img = cv2.cvtColor(final_hsv, cv2.COLOR_HSV2BGR)
    
    return img

def change_sat(img, value):
    hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
    h, s, v = cv2.split(hsv)

    lim = 255 - value
    s[s > lim] = 255
    s[s <= lim] = np.add(s[s <= lim], value, out=s[s <= lim], casting='unsafe')

    final_hsv = cv2.merge((h, s, v))
    img = cv2.cvtColor(final_hsv, cv2.COLOR_HSV2BGR)
    
    return img 

def change_hue(img, value):
    hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
    h, s, v = cv2.split(hsv)

    lim = 179 - value
    h[h > lim] = 179
    h[h <= lim] = np.add(h[h <= lim], value, out=h[h <= lim], casting='unsafe')

    final_hsv = cv2.merge((h, s, v))
    img = cv2.cvtColor(final_hsv, cv2.COLOR_HSV2BGR)
    
    return img 

def change_angle(img, angle, scaling=1):
    rows, cols, _ = img.shape
    
    M = cv2.getRotationMatrix2D((cols/2,rows/2), angle, scaling)
    dst = cv2.warpAffine(img,M,(cols,rows))
    
    return dst

def gaussian_blur(img, size, sigma=0):
    blur = cv2.GaussianBlur(prac_img,(size,size),sigma)
    
    return blur
```
